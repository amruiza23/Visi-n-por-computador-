```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
im = np.zeros((500, 500), np.uint8)

for i in range(500):
    im[i, i] = 255

plt.figure(figsize=(6, 6))
plt.imshow(im, cmap='gray', interpolation='nearest')
plt.axis('off')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(6, 6))


colors = ['red', 'green', 'blue', 'black']
size = 0.5
for i, color in enumerate(colors):
    ax.add_patch(plt.Rectangle((-size + i*0.1, -size + i*0.1),
                               2*size - 2*i*0.1, 2*size - 2*i*0.1,
                               color=color))

ax.set_xlim(-1, 1)
ax.set_ylim(-1, 1)
ax.set_aspect('equal')
ax.set_facecolor('black')
ax.axis('on')

plt.show()
```
